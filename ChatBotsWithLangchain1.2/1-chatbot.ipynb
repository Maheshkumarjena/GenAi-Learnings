<h4>message history to understand the session management and how the model is able to remember the context of the conversation and respond accordingly. </h4>

<p># -- we can use Message History class to wrap our model and keep it stateful and maintain the context of the conversation.
# -- we can also use the message history to analyze the conversation and understand how the model is responding to different inputs and how it is able to maintain the context of the conversation over multiple turns.</p>

```python
from langchain_community.chat_message_histories import ChatMessageHistory 
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

store={}
def get_session_history(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store or not isinstance(store[session_id], BaseChatMessageHistory):
        store[session_id] = ChatMessageHistory() 
    return store[session_id]

with_message_history = RunnableWithMessageHistory(model ,get_session_history) 

```

```python
config = {
    "configurable": {
            "session_id" : "chat1",
    }
}

```

```python
with_message_history.invoke(
    [HumanMessage(content="Hi , my name is Mahesh and I'm a Software Developer /  AI engineer. What can you do ? ")],
    config = config 
)
```

```python
from langchain_core.messages import SystemMessage, HumanMessage, trim_messages
import tiktoken

encoding = tiktoken.get_encoding("cl100k_base")

def token_counter(messages):
    return sum(len(encoding.encode(m.content)) for m in messages)

trimmer = trim_messages(
    max_tokens=60,
    strategy="last",
    token_counter=token_counter,
    include_system=True,
    allow_partial=False,
    start_on="human",
)

messages = [
    SystemMessage(content="you are a helpful assistant that helps to answer questions to the best of your ability: {language}."),
    HumanMessage(content="Hi My Name is Mahesh"),
    SystemMessage(content="Hi Mahesh, How can I help you?"),
    HumanMessage(content="what is my name and profession?"),
    SystemMessage(content="Your name is Mahesh and you are a Software Engineer"),
    HumanMessage(content="what's the weather like today?"),
    SystemMessage(content="Sorry I don't have access to real time information about weather."),
    HumanMessage(content="Okay, can you tell me a joke?"),
    SystemMessage(content="Sure, Here's a joke for you: Why do programmers prefer dark mode? Because light attracts bugs!"),
]
trimmer.invoke(messages)
```

```python
from operator import itemgetter
from langchain_core.runnables import RunnablePassthrough

chain = (
    RunnablePassthrough.assign(messages=itemgetter("messages"))
    | trimmer
    | prompt
    | model
)

response = chain.invoke({
    "messages": messages + [HumanMessage(content="what is my name?")],
    "language": "English"
}).content
response
```
