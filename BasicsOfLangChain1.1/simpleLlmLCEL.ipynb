```python
### OPEN AI API KEY and open Source models--Llama3 , Gemma2 , mistral--Groq.
import os 
from dotenv import load_dotenv
load_dotenv()

from langchain_ollama import ChatOllama

llm = ChatOllama(
    model="gemma:2b",
    temperature=0.7
)
print(llm)

import openai
groq_api_key = os.getenv("GROQ_API_KEY")

```

```python
from langchain_groq import ChatGroq
model = ChatGroq(model = "llama-3.1-8b-instant", groq_api_key = groq_api_key)
model
```

```python
from langchain_core.messages import HumanMessage , SystemMessage 

messages = [
    SystemMessage(content = "Translate the following from English to French"),
    HumanMessage(content = "Hello , How are you ?")
]

result = model.invoke(messages)
result
```

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
parser.invoke(result)
```

```python
### using LCEL - chain the components (The model has both the messages , the parser and the result . it gets the result then parses it and gives the final output in text format.  )
chain = model | parser 
chain.invoke(messages)
```

```python
from langchain_core.prompts import ChatPromptTemplate 

generic_template = "Translate the following text into {language}:"
prompt = ChatPromptTemplate.from_messages(
    [("system", generic_template), ("user", "{text}")]
)

response=prompt.invoke({"language" : "French" , "text" : "Hello, How are you ?"})
response.to_messages()
```

```python
# chaining the prompt , model and the parser together to get the final output in text format.
chain = prompt | model | parser  
chain.invoke({"language" : "Hindi" , "text" : "Hello, how are you ?"})  
```
